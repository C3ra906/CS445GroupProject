Imports

```python
import numpy as np
from sklearn.model_selection import train_test_split

from dataset import Dataset
from network import NeuralNetwork
from oversample import oversample
```

Load normalized data and split into training and test sets

```python
file = np.loadtxt(open("normalized_data.csv", "rb"), delimiter=",", skiprows=1)

train, test = train_test_split(file, test_size=.3)
oversampled_data = oversample(train)
```

```python
train = Dataset(train)
oversampled_train = Dataset(oversampled_data)
test = Dataset(test)
```

Run Tests

Original input data, no oversampling, threshold of 0.5, learning rate of 0.001, batch_size of 8

```python
network = NeuralNetwork(20, training_data=train,
                        test_data=test, momentum=0, learning_rate=.001, threshold=.5,
                        batch_size=8)
tpr, fpr = network.run()
print(f"True Positive Rate: {tpr}")
print(f"False Positive Rate: {fpr}")
```

Now, use oversampled data with a equal distribution of strokes and non-strokes

```python
network = NeuralNetwork(20, training_data=train,
                        test_data=test, momentum=0, learning_rate=.001, threshold=.5,
                        batch_size=8)
tpr, fpr = network.run()
print(f"True Positive Rate: {tpr}")
print(f"False Positive Rate: {fpr}")
```

Now varying threshold, with original data

```python
for threshold in [.05, .1, .2, .3, .4, .5]:
    network = NeuralNetwork(20, training_data=train,
                            test_data=test, momentum=0, learning_rate=.001, threshold=threshold,
                            batch_size=8)
    tpr, fpr = network.run()
    print(f"True Positive Rate: {tpr}")
    print(f"False Positive Rate: {fpr}")
```

Now varying batch size

```python
for batch_size in [1, 3, 5, 8, 10]:
    network = NeuralNetwork(20, training_data=train,
                            test_data=test, momentum=0, learning_rate=.001, threshold=.05,
                            batch_size=batch_size)
    tpr, fpr = network.run()
    print(f"True Positive Rate: {tpr}")
    print(f"False Positive Rate: {fpr}")
```

Now varying n

```python
for n in [3, 5, 10, 20]:
    network = NeuralNetwork(n, training_data=train,
                            test_data=test, momentum=0, learning_rate=.001, threshold=.05,
                            batch_size=8)
    tpr, fpr = network.run()
print(f"True Positive Rate: {tpr}")
print(f"False Positive Rate: {fpr}")
```
